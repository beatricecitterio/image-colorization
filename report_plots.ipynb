```python
def plot_metrics_by_metric(results, metric):
    loss_types = ['mse_loss', 'l1_loss', 'l1smooth_loss']
    cnn_means = [results['cnn'][loss][metric] for loss in loss_types]
    cnn_stds = [results['cnn'][loss][f'STD {metric.split()[-1]}'] if metric != 'FID' else 0 for loss in loss_types]
    unet_means = [results['unet'][loss][metric] for loss in loss_types]
    unet_stds = [results['unet'][loss][f'STD {metric.split()[-1]}'] if metric != 'FID' else 0 for loss in loss_types]
    x = np.arange(len(loss_types))  # the label locations

    if metric == 'FID':
        plt.plot(x - 0.1, cnn_means, 'o', color=cnn_color_test, label='CNN')
        plt.plot(x + 0.1, unet_means, 'o', color=unet_color_test, label='U-Net')
        plt.xticks(fontsize=13)
        plt.yticks(fontsize=12)
        plt.title(f'{metric}', fontsize=14)
        plt.xticks(x, ['MSE', '$\mathcal{L}_1$', '$\mathcal{L}_1$-smooth'])
        plt.grid(True)
    else:
        plt.errorbar(x - 0.1, cnn_means, yerr=cnn_stds, fmt='o', color=cnn_color_test, label='CNN', capsize=5)
        plt.errorbar(x + 0.1, unet_means, yerr=unet_stds, fmt='o', color=unet_color_test, label='U-Net', capsize=5)
        plt.xticks(fontsize=13)
        plt.yticks(fontsize=10)
        plt.title(f'{metric} $\pm$ std', fontsize=14)
        plt.xticks(x, ['MSE', '$\mathcal{L}_1$', '$\mathcal{L}_1$-smooth'])
        plt.grid(True)

        # Adjust the positions of the tick labels to avoid overlap
        ax = plt.gca()
        xtick_labels = ax.get_xticklabels()
        for i, label in enumerate(xtick_labels):
            if i == 0:
                label.set_horizontalalignment('left')
            elif i == len(xtick_labels) - 1:
                label.set_horizontalalignment('right')
            else:
                label.set_horizontalalignment('center')

metrics = ['Average MSE', 'Average PSNR', 'Average SSIM', 'FID']
fig, ax = plt.subplots(figsize=(12, 10))

# Remove x and y axis ticks and labels from the overall plot
plt.tick_params(labelbottom=False, labelleft=False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

for i, metric in enumerate(metrics, 1):
    ax = plt.subplot(2, 2, i)
    plot_metrics_by_metric(results, metric)

plt.tight_layout()

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.05), fontsize=13)
plt.show()
```
