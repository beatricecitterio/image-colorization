```python
results = {
    'cnn': {
        'mse_loss': {
            'Average MSE': 0.0034, 'STD MSE': 0.0004,
            'Average PSNR': 49.4572, 'STD PSNR': 1.0570,
            'Average SSIM': 0.8129, 'STD SSIM': 0.1061,
            'FID': 8.9813
        },
        'l1_loss': {
            'Average MSE': 0.0033, 'STD MSE': 0.0004,
            'Average PSNR': 49.6630, 'STD PSNR': 1.0990,
            'Average SSIM': 0.8143, 'STD SSIM': 0.1066,
            'FID': 8.9523
        },
        'l1smooth_loss': {
            'Average MSE': 0.0034, 'STD MSE': 0.0005,
            'Average PSNR': 49.5062, 'STD PSNR': 1.1490,
            'Average SSIM': 0.8145, 'STD SSIM': 0.1067,
            'FID': 9.0082
        }
    },
    'unet': {
        'mse_loss': {
            'Average MSE': 0.0026, 'STD MSE': 0.0003,
            'Average PSNR': 51.6446, 'STD PSNR': 1.1107,
            'Average SSIM': 0.8173, 'STD SSIM': 0.0919,
            'FID': 7.5709
        },
        'l1_loss': {
            'Average MSE': 0.0027, 'STD MSE': 0.0004,
            'Average PSNR': 51.3193, 'STD PSNR': 1.1742,
            'Average SSIM': 0.8193, 'STD SSIM': 0.0969,
            'FID': 7.9728
        },
        'l1smooth_loss': {
            'Average MSE': 0.0026, 'STD MSE': 0.0003,
            'Average PSNR': 51.6279, 'STD PSNR': 1.1127,
            'Average SSIM': 0.8165, 'STD SSIM': 0.0916,
            'FID': 7.4200
        }
    }
}

def plot_metrics(results, loss_type):
    metrics = ['Average MSE', 'Average PSNR', 'Average SSIM', 'FID']
    cnn_means = [
        results['cnn'][loss_type]['Average MSE'],
        results['cnn'][loss_type]['Average PSNR'],
        results['cnn'][loss_type]['Average SSIM'],
        results['cnn'][loss_type]['FID']
    ]
    cnn_stds = [
        results['cnn'][loss_type]['STD MSE'],
        results['cnn'][loss_type]['STD PSNR'],
        results['cnn'][loss_type]['STD SSIM'],
        0  # FID does not have a standard deviation 
    ]
    unet_means = [
        results['unet'][loss_type]['Average MSE'],
        results['unet'][loss_type]['Average PSNR'],
        results['unet'][loss_type]['Average SSIM'],
        results['unet'][loss_type]['FID']
    ]
    unet_stds = [
        results['unet'][loss_type]['STD MSE'],
        results['unet'][loss_type]['STD PSNR'],
        results['unet'][loss_type]['STD SSIM'],
        0  # FID does not have a standard deviation
    ]

    x = np.arange(len(metrics))  # the label locations

    plt.figure(figsize=(10, 6))
    plt.errorbar(x - 0.1, cnn_means, yerr=cnn_stds, fmt='o', color='#a00000', label='CNN', capsize=5)
    plt.errorbar(x + 0.1, unet_means, yerr=unet_stds, fmt='o', color='#1a80bb', label='U-Net', capsize=5)

    plt.ylabel('Scores')
    plt.title(f'Metrics by model and loss type ({loss_type})')
    plt.xticks(x, metrics)
    plt.legend()

    plt.grid(True)

    plt.tight_layout()
    plt.show()

for loss_type in results['cnn'].keys():
    plot_metrics(results, loss_type)
```
