```python
cnn_l1_train, cnn_l1_test = load_losses("losses/cnn_l1.txt")
cnn_l1smooth_train, cnn_l1smooth_test = load_losses("losses/cnn_l1smooth.txt")
unet_mse_train, unet_mse_test = load_losses("losses/unet_mse.txt")
unet_l1_train, unet_l1_test = load_losses("losses/unet_l1.txt")
unet_l1smooth_train, unet_l1smooth_test = load_losses("losses/unet_l1smooth.txt")
```

```python
epochs = list(range(1, 21))
plt.figure(figsize=(10, 6))

cnn_color = '#0072B2'  # Blue
unet_color = '#D55E00'  # Reddish-brown

plt.plot(epochs, cnn_mse_train, '--', linewidth=1.5, color=cnn_color, label='CNN Train Loss')
plt.plot(epochs, cnn_mse_test, linewidth=2, color=cnn_color, label='CNN Test Loss')
plt.plot(epochs, unet_mse_train, '--', linewidth=1.5, color=unet_color, label='U-Net Train Loss')
plt.plot(epochs, unet_mse_test, linewidth=2, color=unet_color, label='U-Net Test Loss')

plt.xlabel('Epochs', fontsize=13)
plt.ylabel('Loss', fontsize=13)
plt.grid(linewidth=0.5)
plt.legend(loc='best', fontsize=12)
plt.xticks(epochs, fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(epochs)
plt.title("MSE Loss")
plt.tight_layout()
plt.show()
```
